```python
import pandas as pd

df_data = pd.read_csv('../../Data/Input/Census-Data/GreaterLondon_2011_OAC_Raw_uVariables--zscores-with-colnames.csv')
df_data
```

```python
df_data_drop_first_column = df_data.iloc[:,1:]
```

```python
from sklearn.decomposition import PCA

reduced_data = PCA(n_components=60).fit_transform(df_data_drop_first_column.values)
reduced_data
```

```python
reduced_df = pd.DataFrame(reduced_data, columns = ["pyPC"+str(i+1) for i in range(60)])
data_with_reduced_df = pd.concat([df_data, reduced_df.reindex(df_data.index)], axis=1)
data_with_reduced_df
```

```python
data_with_reduced_df.to_csv("../../Data/Input/Census-Data/GreaterLondon_2011_OAC_Raw_uVariables--zscores-to-pyPCA60.csv", index=False)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8, random_state=0).fit(reduced_data)
df_labels = pd.DataFrame(kmeans.labels_)
df_labels.columns = ['pyPCA60kmeans']
#Convert the clusters created by K-Means to csv file
data_with_labels_df = pd.concat([df_data.iloc[:,:1], df_labels.reindex(df_data.index)], axis=1)
data_with_labels_df.to_csv("../../Data/Output/GeoCMeans/oac_pyPCA60_kmeans.csv", index=False)
```
